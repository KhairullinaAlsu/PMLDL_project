```python
columns_to_drop = ['track_album_release_date', 'playlist_name', 'track_album_name', 'track_album_id']
tracks_cleaned = tracks.drop(columns=columns_to_drop)
```

```python
from sklearn.preprocessing import StandardScaler

numeric_features = ['track_popularity', 'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms']

scaler = StandardScaler()
tracks_cleaned[numeric_features] = scaler.fit_transform(tracks_cleaned[numeric_features])
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
tracks_cleaned[numeric_features] = scaler.fit_transform(tracks_cleaned[numeric_features])
```

```python
from sklearn.preprocessing import LabelEncoder

categorical_features = ['artistname', 'track_id', 'playlist_id', 'playlist_genre', 'playlist_subgenre']

for feature in categorical_features:
    le = LabelEncoder()
    tracks_cleaned[feature] = le.fit_transform(tracks_cleaned[feature])
```

```python
tracks_for_predictions = tracks_cleaned[['trackname']]
tracks_for_predictions.head()
```

```python
users_tracks = pd.read_csv('data/user_favorite_tracks.csv') 
users_tracks.head()
```

```python
users_tracks_cleaned = users_tracks.dropna()
users_tracks_cleaned.isnull().sum()
```

```python
label_encoder_names = LabelEncoder()

tracks_cleaned['trackname'] = label_encoder_names.fit_transform(tracks_cleaned['trackname'])

tracks_for_predictions['trackname'] = label_encoder_names.transform(tracks_for_predictions['trackname'].values)

favorite_songs = [f'favorite_song_{i}' for i in range(1, 21)]
for song in favorite_songs:
    users_tracks_cleaned[song] = label_encoder_names.transform(users_tracks_cleaned[song].values)

users_tracks_cleaned['user_id'] = users_tracks_cleaned.index + 1
users_tracks_cleaned = users_tracks_cleaned.drop(columns='user')

cols = users_tracks_cleaned.columns.tolist()
cols = ['user_id'] + cols[:-1]
users_tracks_final = users_tracks_cleaned[cols]

users_tracks_final.head()
```

```python
tracks_for_predictions.to_csv('data/tracks_for_predictions.csv', index=False)
tracks_cleaned.to_csv('data/cleaned_tracks_dataset.csv', index=False)
users_tracks_final.to_csv('data/users_tracks_cleaned.csv', index=False)
```
